Data Cleaning

```python
# All value in column notes and status is missing
df.dropna(subset=['Notes','Status'])
```

```python
Mean_BasePay = df.BasePay.mean()
Mean_OvertimePay = df.OvertimePay.mean()
Mean_OtherPay = df.OtherPay.mean()
Mean_Benefits = df.Benefits.mean()


df['BasePay'].fillna(Mean_BasePay)
df['OvertimePay'].fillna(Mean_OvertimePay)
df['OtherPay'].fillna(Mean_OtherPay)
df['Benefits'].fillna(Mean_Benefits)
```

```python
df.head()
```

Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['TotalPayBenefits'], bins=10)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Histogram: Distribution of Salaries')
plt.show()
```

```python
department_counts = df['JobTitle'].value_counts()
plt.bar(department_counts.index, department_counts.values)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Bar Chart: Proportion of Employees in Different Departments')
plt.xticks(rotation=30)
plt.show()
```

Grouped Analysis

```python
grouped_data = df.groupby(['TotalPayBenefits', 'TotalPay'])['TotalPay'].agg(['mean', 'median', 'min', 'max', 'std'])
average_salaries = grouped_data['mean']
average_salaries
```

```python
grouped_data = df.groupby(['Year'])['TotalPay'].agg(['mean', 'median', 'min', 'max', 'std'])
average_salaries = grouped_data['mean']
average_salaries
```

Simple Correlation Analysis

```python

corr = df['TotalPayBenefits'].corr(df['TotalPay'])

# Plotting a scatter plot
plt.scatter(df['TotalPay'], df['TotalPayBenefits'])
plt.xlabel('TotalPay')
plt.ylabel('Salary')
plt.title('Salary vs TotalPay ')
plt.show()
```
